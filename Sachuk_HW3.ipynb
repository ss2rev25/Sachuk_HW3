```python
import random
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("c:/Users/ivan/Desktop/Unik/КСВД/Сачук_HW3/spotify-2023.csv", encoding="latin-1")
data["streams"].astype("int")
data
```

1. Завдання полягає в тому, щоб побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань.

```python
top10 = data[["track_name","streams"]].sort_values(by="streams").tail(10)
plt.barh(top10["track_name"], top10["streams"])
plt.title("Top 10 most streamed songs")
plt.xlabel("Number of listenings")
plt.ylabel("Song name")
plt.show()
```

4. Рейтинг артистів на Spotify: Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших артистів на Spotify у 2023 році за кількістю слухачів.

```python
ex4 = data.groupby("artist(s)_name")["in_spotify_charts"].sum().sort_values(ascending=False).head(10)
plt.barh(ex4.index, ex4.values)
plt.xlabel("in spotify charts")
plt.ylabel("Artists")
plt.title("Top 10 artists in spotify charts")
```
